# Theory of Nearest Neighbor Methods

Nearest Neighbor Methods:

"Find the most similar image to the given image $x$."

Basic:

- distance / dissimilarity function $d(x,y)$
- database $D = \\{y_1,...,y_n\\}$
- retrieval $ \xi = \arg\min_i d(x,y_i) $

Ranked retrieval:

- retrieve the $k$ nearest neighbors
- rank by dissimilarity

Nearest Neighbor Methods vs Classification:

Problems:

- assume we have a classification problem $p(c,x)$
- classes $c$, measurements $x$


Nearest neighbor algorithms:

- nearest neighbor $\tilde{D}(x) = c_i$ where $i = \arg\min_i d(x,y_i)$
- $k$-nearest neighbor is majority vote of class among $k$ nearest neighbors


Nearest Neighbor Methods vs Classification Error:


Bayes error:

- optimal classification (0-1 loss) is given by $D(x) = \arg\max_c P(c|x)$

Nearest Neighbor error:

- nearest neighbor classifier has no more than twice Bayes error rate
- as $k \rightarrow \infty$, the $k$-NN error approaches the Bayes error

Retrieval as Classification:

- retrieval can be viewed as classification:
- two-class problem: relevant vs not relevant, $P(\hbox{relevant} | x , y_i)$

We don't need to make the classes explicit: nearest neighbors are likely to have the same label(s) as the unknown sample, with the same guarantees as nearest neighbor classifiers.

Image Labeling as Classification:

- each image $y_i$ in the database is classified into one or more classes $c^j_i$ by one or more classifiers
- no matter what the classifier, if $y_i$ is a nearest neighbor of $x$, then $x$ probably has similar class labels
- both 1-NN and $k$-NN results apply
- beauty of $k$-NN vs other classifiers: we don't need to train separate classifiers for each class

# Applying Nearest Neighbor Methods

Applying Nearest Neighbor Methods:

Input:

- measurement vectors $x$

We need to choose:

- feature extraction: $f(x)$ (for now, identity function)
- distance / dissimilarity function: $d(x,y)$

Dissimilarity Functions:

- $d(x,y)$ gets bigger the more "dissimilar" $x$ and $y$ are
- if $x$ and $y$ are vectors, we often use metrics:
  - $d(x,x) = 0$
  - $d(x,y) = d(y,x)$
  - $d(x,y) \geq 0$
  - $d(x,y)+d(y,z) \geq d(x,z)$ (triangle inequality)


Common Dissimilarity Functions:

- $l$-norm: $d(x,y) = (\sum |x_i-y_i|^l)^{1/l}$
- $0$-norm: $d(x,y) = \sum \left\lfloor x_i \neq y_i \right\rfloor$
- $\infty$-norm: $d(x,y) = \max_i |x_i-y_i|$
- cosine-dissimilarity: $d(x,y) = \frac{x\cdot y}{||x|| ~ ||y||}$

# Nearest Neighbor Tools

Libraries:

Two primary libraries for dealing with nearest neighbor problems:

- `FLANN` - fast approximate nearest neighbor library

We'll return to FLANN later, let's look at the `distance` package for now.

```python
hdf = tables.openFile("sample.h5","r")
```

We generally represent datasets as 2D arrays, with each row being a feature vector or sample.



```python
n = len(hdf.root.icons)
data = zeros((n,prod(hdf.root.icons[0].shape)))
for i in range(n):
    data[i,:] = hdf.root.icons[i].ravel()
```

```python
from scipy.spatial import distance
ds = distance.cdist(data[:10],data)
ds[7,:10]
```

```python
argsort(ds[7])[:10]
```

```python
figsize(12,8)
for i,j in enumerate(argsort(ds[7])[:6]):
    subplot(1,6,i+1); axis("off")
    imshow(hdf.root.icons[j])
```

```python
def test(metric='euclidean',p=2,image=7):
    ds = distance.cdist(data[image:image+1],data,metric=metric,p=p)
    for i,j in enumerate(argsort(ds[0])[:6]):
        subplot(1,6,i+1); axis("off")
        imshow(hdf.root.icons[j])
test()
```

```python
test(metric='cityblock')
```

```python
test(metric='chebyshev')
```

```python
test(metric='minkowski',p=4)
```

```python
test(metric='minkowski',p=.2)
```

# Feature Extraction

```python
def test(metric='euclidean',p=2,image=7,f=lambda x:x):
    n = len(hdf.root.icons)
    d = prod(f(hdf.root.icons[0]).shape)
    data = zeros((n,d))
    for i in range(n):
        data[i,:] = f(hdf.root.icons[i]).ravel()
    ds = distance.cdist(data[image:image+1],data,metric=metric,p=p)
    for i,j in enumerate(argsort(ds[0])[:6]):
        subplot(1,6,i+1); axis("off")
        imshow(hdf.root.icons[j])
test()
```

```python
def f(image): return interpolation.zoom(image,(1/16.0,1/16.0,1))
imshow(f(hdf.root.icons[7]),interpolation='nearest')
```

```python
test(f=f)
```

```python
def f(image): return filters.gaussian_filter(image,(10,10,0))
imshow(f(hdf.root.icons[7]),interpolation='nearest')
```

```python
test(f=f)
```

```python
def f(image):
    image = mean(image,axis=2)
    image = filters.gaussian_gradient_magnitude(image,5.0)
    return image
imshow(f(hdf.root.icons[7]))
```

```python
test(f=f)
```

```python
def f(image):
    image = array(mean(image,axis=2),'f')
    hist = cv2.calcHist([image],[0],None,[256],[0,256])
    return hist[1:]
plot(f(hdf.root.icons[7]))
```

```python
test(f=f,metric='braycurtis')
```
