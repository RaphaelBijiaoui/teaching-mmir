```python

matplotlib.rc("image",cmap="gray")
from scipy.ndimage import filters
from scipy.ndimage.filters import gaussian_filter as G
def crop_black(image):
    gray = image
    if gray.ndim==3: gray = sum(image,axis=2)
    yr = find(sum(gray,axis=1)>0)
    y0 = yr[0]
    y1 = yr[-1]
    xr = find(sum(gray,axis=0)>0)
    x0 = xr[0]
    x1 = xr[-1]
    if image.ndim==3: return image[y0:y1,x0:x1,:]
    else: return image[y0:y1,x0:x1]
```

# Idea behind Harris

Corners / interest points:

- corners are places where edges of different orientations meet
- interest points are places that can be detected stably

## Stability

Stable corner locations:

Let's define stability as the difference between the shifted image and the image, averaged over a window $w$.

$$ S(x,y) = \sum_u \sum_v w(u,v) \left( I(u+x,v+y) - I(x,y)\right)^2 $$

$I(u+x,v+y)$ can be approximated by a Taylor series:

$$ I(u+x,v+y) \approx I(u,v) + I_x(u,v)x+I_y(u,v)y $$

This then gives rise to:

$$ S(x,y) \approx \sum_u \sum_v w(u,v) \left( I_x(u,v)x + I_y(u,v)y \right)^2 $$

Matrix expression:

$$ S(x,y) \approx \sum_u \sum_v w(u,v) \, \left( I_x(u,v)x + I_y(u,v)y \right)^2 $$

This can be written in matrix form:

$$ S(x,y) \approx \begin{pmatrix} x & y \end{pmatrix} A \begin{pmatrix} x \\\\ y \end{pmatrix} $$

$$
A = \sum_u \sum_v w(u,v) 
\begin{bmatrix}
I_x^2 & I_x I_y \\\\
I_x I_y & I_y^2 
\end{bmatrix}
$$

That derivation is fine, but another derivation is perhaps more straightforward.

The matrix we are computing the eigenvalues of is similar to the covariance matrix (we already know that the "mean" should be at (0,0)):

$$ A = \pmatrix { \sum_i (g^x_i)^2 & \sum_i g^x_i g^y_i \\\\  \sum_i g^x_i g^y_i & \sum_i (g^y_i)^2 } $$

Computing this over a patch is the same as summing $g^x$ and $g^y$ locally with a uniform filter.
However, we can generalize this and use a Gaussian filter, giving smoother estimates.



```python
dx = filters.gaussian_filter(corner,1,(0,1))
dy = filters.gaussian_filter(corner,1,(1,0))
sxx = filters.gaussian_filter(dx**2,5.0)
sxy = filters.gaussian_filter(dx*dy,5.0)
syy = filters.gaussian_filter(dy**2,5.0)
```

```python
subplot(131); imshow(sxx)
subplot(132); imshow(sxy)
subplot(133); imshow(syy)
```

Eigenvalue computation:

The eigenvalue computation requires solving:

$$ \left(A - \lambda I\right) v = 0 $$

In the $2\times 2$ case:

$$ {\rm det} \pmatrix{ \lambda - a & -b \\\\ -c & \lambda - d } = 0 $$

This gives rise to the quadratic equation:

$$\lambda^2 - \left( a + d \right )\lambda + \left ( ad - bc \right ) = \lambda^2 - \lambda {\rm tr}(A) + {\rm det}(A) = 0 $$

The solution is given by:

$$ \lambda = \frac{{\rm tr}(A) \pm \sqrt{{\rm tr}^2 (A) - 4 {\rm det}(A)}}{2} $$

```python
# Eigenvalue computation
T = sxx+syy
D = sxx*syy-sxy**2
R = (T**2-4*D)**.5
hi = (T+R)/2
lo = (T-R)/2
subplot(121); imshow(hi)
subplot(122); imshow(lo)
```

```python
def harris(image,sigma=1.0,area=5.0):
    dx = filters.gaussian_filter(image,sigma,(0,1))
    dy = filters.gaussian_filter(image,sigma,(1,0))
    sxx = filters.gaussian_filter(dx**2,area)
    sxy = filters.gaussian_filter(dx*dy,area)
    syy = filters.gaussian_filter(dy**2,area)
    T = sxx+syy
    D = sxx*syy-sxy**2
    R = (T**2-4*D)**.5
    hi = (T+R)/2
    lo = (T-R)/2
    return lo
```

```python
# Harris corner detector
subplot(131); imshow(harris(flat),vmax=0.002)
subplot(132); imshow(harris(edge),vmax=0.002)
subplot(133); imshow(harris(corner),vmax=0.002)
```
