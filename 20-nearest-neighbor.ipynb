```python
from scipy.ndimage import interpolation
import tables
```

```python
hdf = tables.openFile("sample.h5","r")
```

```python
for i in range(9):
    subplot(3,3,i+1); axis("off")
    imshow(hdf.root.icons[i])
```

```python
n = len(hdf.root.icons)
data = zeros((n,prod(hdf.root.icons[0].shape)))
for i in range(n):
    data[i,:] = hdf.root.icons[i].ravel()
```

```python
from scipy.spatial import distance
ds = distance.cdist(data[:100:10],data)
figsize(12,8)
for j in range(6):
    nn = argsort(ds[j])
    for i in range(6):
        subplot(6,6,j*6+i+1); axis("off")
        imshow(hdf.root.icons[nn[i]])
```
